```python

import csv
import random

# Real-world Entity Pools
personalities = [
    "Lionel Messi", "Cristiano Ronaldo", "LeBron James", "Serena Williams", "Virat Kohli",
    "Barack Obama", "Winston Churchill", "Abraham Lincoln", "Nelson Mandela", "Mahatma Gandhi"
]
monuments = [
    "Eiffel Tower", "Taj Mahal", "Statue of Liberty", "Colosseum", "Great Wall of China",
    "Empire State Building", "Machu Picchu", "Sydney Opera House", "Burj Khalifa", "Big Ben"
]
cities = ["Paris", "Agra", "New York", "Rome", "Beijing", "London", "Cairo", "Tokyo", "Sydney", "Dubai"]
countries = ["France", "India", "USA", "Italy", "China", "UK", "Egypt", "Japan", "Australia", "UAE"]
sports_teams = ["Real Madrid", "Los Angeles Lakers", "Manchester United", "New York Yankees", "Ferrari F1 Team"]

triples_set = set()

# 1. Core Semantic Baseline Relationships
for p in personalities:
    triples_set.add((p, "is_a", "Famous Personality"))
for m in monuments:
    triples_set.add((m, "is_a", "Historical Monument"))

# 2. Scale up to exactly 5,000 unique links (Relationships + Subject Attributes)
counter = 1
while len(triples_set) < 5000:
    # Anchor entities using index patterns or randomized variations
    person = random.choice(personalities)
    monument = random.choice(monuments)
    city = random.choice(cities)
    country = random.choice(countries)
    team = random.choice(sports_teams)

    # Dynamic synthetic variations to avoid duplicates and reach 5k limit smoothly
    var_person = f"{person}_FanClub_{counter % 200}"
    var_monument = f"{monument}_Section_{counter % 200}"

    # --- Real-World Semantic Relationships ---
    triples_set.add((person, "associated_with_team", team))
    triples_set.add((monument, "located_in_city", city))
    triples_set.add((city, "part_of_country", country))
    triples_set.add((person, "visited_monument", monument))

    # --- Subject Attributes (Represented as Object Literals) ---
    if len(triples_set) < 5000:
        triples_set.add((person, "has_net_worth_millions", str(random.randint(50, 1200))))
    if len(triples_set) < 5000:
        triples_set.add((monument, "constructed_in_year", str(random.randint(80, 2010))))
    if len(triples_set) < 5000:
        triples_set.add((city, "has_safety_rating", random.choice(["High", "Medium", "Excellent"])))

    # --- Structural Padding using Valid Graph Topology ---
    if len(triples_set) < 5000:
        triples_set.add((var_person, "follows_career_of", person))
    if len(triples_set) < 5000:
        triples_set.add((var_monument, "belongs_to_complex", monument))

    counter += 1

# Flatten set to list and trim any accidental boundary overflow
final_triples = list(triples_set)[:5000]

# 3. Write data to the CSV file
filename = "dbpedia.csv"
headers = ["subject", "predicate", "object"]

with open(filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(final_triples)

print(f"File created successfully: '{filename}'")
print(f"Total rows written: {len(final_triples)} graph rows.")
```
